```python
import pandas as pd
```

```python
datos= pd.read_csv('dataframe_tratado.csv')
```

```python
datos.drop(columns={'Unnamed: 0'},inplace=True)
```

```python
datos.head()
```

```python
datos['Churn']= datos['Churn'].astype('category')
```

```python
datos['SeniorCitizen']= datos['SeniorCitizen'].astype('category')
datos['Partner']= datos['Partner'].astype('category')
datos['Dependents']= datos['Dependents'].astype('category')
```

```python
datos.info()
```

```python
columnas = datos.select_dtypes(include='object').columns.tolist()
columnas
```

```python
df=pd.get_dummies(data=datos, columns=columnas, dtype='category')
```

```python
df.info()
```

```python
columnas_a_convertir = df.select_dtypes(include='category').columns.tolist()
df = df.astype({col: 'int64' for col in columnas_a_convertir})
```

```python
#hacemos el balanceo de clases
!pip install imblearn
from imblearn.over_sampling import SMOTE
```

```python
X= df.drop(columns=['Churn'])
y= df['Churn']
```

```python
oversampling = SMOTE(random_state=42)
X_balanced, y_balanced = oversampling.fit_resample(X, y)
```

```python
#comprobamos que el balanceo de clases se haya realizado correctamente
y_balanced.value_counts(normalize=True)
```

```python
#visualizamos la matriz de correlacion de los datos balanceados
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df_balanced = pd.concat([X_balanced, y_balanced], axis=1)
plt.figure(figsize=(20, 20))
correlation_matrix = df_balanced.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Matriz de Correlación de Datos Balanceados')
plt.show()
```

```python
#ordenamos la correlacion ascendentemente con respecto a Churn
correlation_churn = df_balanced.corr()['Churn'].sort_values(ascending=False)
print(correlation_churn)
```

Las correlaciones mas fuertes son con Contract_Month-to-month, InternetService_Fiber optic, OnlineSecurity_No, TechSupport_No y relaciones inversas con tenure, Contract_two year y Online Security yes

#### Modelo Predictivo

```python
df_balanced.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.2, random_state=42)
```

```python
#modelo predictivo 1 regresion logistica para clasificacion binaria de Churn
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
# Creamos el modelo de regresión logística
modelo_1 = LogisticRegression()
# realizamos la normalizacion de los datos
scaler = StandardScaler()
#tener en cuenta que uno es fit_transform y el otro es transform
# fit_transform se usa en el conjunto de entrenamiento
X_train_scaled = scaler.fit_transform(X_train)
# transform se usa en el conjunto de test
X_test_scaled = scaler.transform(X_test)
```

```python
#ahora entrenamos nuestro modelo de regresion logistica
modelo_1.fit(X_train_scaled, y_train)
#ahora hacemos las predicciones
y_pred_1 = modelo_1.predict(X_test_scaled)
```

```python
#Ahora hacemos la clasificacion con otro modelo, por ejemplo un arbol de decision
from sklearn.tree import DecisionTreeClassifier
modelo_2 = DecisionTreeClassifier(max_depth=5, random_state=42)
#aqui no necesitamos normalizar los datos, no es sensible a los outliers
modelo_2.fit(X_train, y_train)
#ahora hacemos las predicciones
y_pred_2 = modelo_2.predict(X_test)
```

```python
#Evaluacion y comparacion de los modelos
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# Evaluación del modelo de regresión logística
print(f"***********Evaluacion del modelo de regresión logística*********** \n {classification_report(y_test, y_pred_1)}")
print(f"***********Evaluacion del modelo de arbol de decision*********** \n {classification_report(y_test, y_pred_2)}")
```

```python
#Evaluamos los modelos con la matriz de confusion
from sklearn.metrics import ConfusionMatrixDisplay
disp_1 =ConfusionMatrixDisplay.from_estimator(modelo_1, X_test_scaled, y_test, cmap='Blues', display_labels=['cancelo','no cancelo'])
disp_2= ConfusionMatrixDisplay.from_estimator(modelo_2, X_test, y_test, cmap='Blues', display_labels=['cancelo', 'no cancelo'])

disp_1.ax_.set_title('Matriz de Confusión - Regresión Logística')
disp_1.ax_.grid(False)
disp_2.ax_.grid(False)
disp_2.ax_.set_title('Matriz de Confusión - Árbol de Decisión')
```

```python
modelo_1.coef_
```

```python
#analisis de importancia de las variables ordenamos las variables de mayor a menor importancia
#para regresion logistica de acuerdo a los coeficientes
import numpy as np
import pandas as pd
coeficientes = pd.Series(modelo_1.coef_[0], index=X_train.columns)
# a partir de los coeficientes hallamos la importancia de las variables
importancias_df = pd.DataFrame(coeficientes, columns=['Coeficientes'])
importancias_df = importancias_df.sort_values(by='Coeficientes', ascending=False)
print(importancias_df)
```

```python
#para el caso de arbol de decision 
importances = modelo_2.feature_importances_
importances_df = pd.DataFrame(importances, index=X_train.columns, columns=['Importancia'])
importances_df = importances_df.sort_values(by='Importancia', ascending=False) 
print(importances_df)
```

#### Conclusiones

**Rendimiento de modelos**

- Regresión logística:   accuracy ≈ 0.84, precision/recall/f1 ≈ 0.84 en ambas clases.

    Matriz de confusión:   TP=898, FN=164, FP=164, TN=840 → detecta más cancelaciones reales y comete menos errores que el árbol.

- Árbol de decisión:   accuracy ≈ 0.80, precision/recall/f1 ≈ 0.80.

Matriz de confusión:   TP=847, FN=215, FP=196, TN=808 → más falsos negativos y falsos positivos.

**Conclusión de performance:**  
La regresión logística es el mejor candidato para producción en tu set actual (mejor recall y f1 en la clase 1).

**Variables que más influyen**

- Month-to-month aparece como el factor más asociado al churn (alta importancia en el árbol).

- One year / Two year muestran efecto protector (coeficientes negativos en logística).

- OnlineSecurity, TechSupport, DeviceProtection, OnlineBackup con valores “Yes”/activos se asocian a menor churn (coeficientes negativos; también con importancia no trivial en el árbol).

- Métodos automáticos (ej. Credit card (automatic), Bank transfer (automatic)) se relacionan con menor churn (coeficientes negativos).

- Total (gasto acumulado) con coeficiente positivo: a mayor gasto total, mayor probabilidad de cancelar (en tu muestra).

- Tenure: aporta (importancia moderada) y suele reducir el riesgo cuanto más tiempo lleva el cliente.


**Estrategias de retencion**

- Ofertas de actualizacion de contrato a los del tipo 1–2 años con beneficios iniciales. Disminuir mes a mes
- Mensajes y retención proactiva a quienes están en mes a mes y con gasto total alto.
- Fomentar servicios de valor agregado
- Paquetes con OnlineSecurity / TechSupport / DeviceProtection como add-ons de fidelización.
- Promover pagos automáticos
- Descuentos o puntos por adhesión a débito/auto-pay.
- Programas de lealtad por antigüedad (tenure)
- Beneficios escalonados que aumenten con el tiempo de permanencia.
